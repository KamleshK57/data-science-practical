```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv("C:/Users/Lenovo/Desktop/Documents/irisdataset.csv")
fig=iris[iris.species=='iris_setosa'].plot.scatter(x='petallength',y='petal width',color='orange',label='setosa')
iris[iris.species=='iris_versicolor'].plot.scatter(x='petallength',y='petal width',color='blue',label='versicolor',ax=fig)
iris[iris.species=='iris_virginica'].plot.scatter(x='petallength',y='petal width',color='green',label='virginica',ax=fig)
fig.set_xlabel("petal length")
fig.set_ylabel("petal width")
fig.set_title("petal length vs width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```
